```python
def clean_link(link):
    # Remove "https://" or "http://" if present
    link = re.sub(r'^https?://', '', link)

    # Remove "www." if present
    link = re.sub(r'^www\.', '', link)
    
    # Remove "www." if present
    link = re.sub(r'^www2\.', '', link)

    # Remove domain extension, e.g., ".com" or ".org"
    link = re.sub(r'\.[a-zA-Z]{2,3}$', '', link)

    # Replace special characters with a space
    link = re.sub(r'[^a-zA-Z\s]', ' ', link)

    # Convert to lowercase
    link = link.lower()

    return link
```

```python
### Stop words removel
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))  # set of English stop words

# function to remove stop words from a text
def remove_stop_words(text):
    words = text.split()
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)
```

```python
df['all'] = df['all'].apply(remove_stop_words)
```

```python
import nltk
nltk.download('punkt')
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')


# create an instance of WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# define a function to lemmatize a given sentence
def lemmatize_sentence(sentence):
    tokens = nltk.word_tokenize(sentence) # tokenizing the string
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return ' '.join(lemmatized_tokens)

```

```python
# apply the lemmatize_sentence function to the text column
df1['lemmatized_text'] = df1['all'].apply(lemmatize_sentence)
```

```python
df1['lemmatized_text'][0]
```

```python
def remove_single_letter_words(text):
    # Remove words if they consist of a single letter
    cleaned_text = re.sub(r'\b\w{1,2}\b', '', text)

    return cleaned_text
```

```python
df1['lemmatized_text'] = df1['lemmatized_text'].apply(remove_single_letter_words)
```
